1. What is the Probability density function (PDF)?

    - The probability density function (PDF) is a function used in probability theory to describe the probability distribution of a continuous random variable.
    - The PDF describes the relative likelihood(possibility) of a value occurring within a given range of values, by integrating the function over that range. 
    - The area under the PDF curve gives the probability of an event occurring within that range.

2. What are the types of Probability distribution?

    There are many types of probability distributions, but some of the most common ones are -
    - Normal distribution (Gaussian distribution) $\to$ PDF
    - Binomial distribution $\to$ PMF
    - Poisson distribution $\to$ PMF
    - Exponential distribution $\to$ PMF
    - **Gamma distribution** $\to$ PDF
    - Uniform distribution $\to$ PMF
    - Log-normal distribution $\to$ PDF

    **Gamma distribution :**
    
    The gamma distribution is a continuous probability distribution that has a probability density function $\to$ (PDF). 
    
    The PDF of the gamma distribution is given by - 

    $f(x) = \frac{(\frac{x-\mu}{\beta})^{\gamma - 1}\exp{(-\frac{x-\mu} {\beta}})} {\beta\Gamma(\gamma)}  \hspace{.2in}  x \ge \mu; \gamma, \beta > 0$

    where $\gamma$ is the shape parameter, $\mu$ is the location parameter, $\beta$ is the scale parameter, and $\Gamma$ is the gamma function which has the formula

    $\Gamma(a) = \int_{0}^{\infty} {t^{a-1}e^{-t}dt}$

    The gamma function is an extension of the factorial function to real (and complex) numbers. Specifically, if $n$ is a positive integer, then

    $\Gamma(n) = (n-1)!$

    The gamma distribution has many applications in statistics and engineering. For example, it can model the waiting time until $k$ events occur in a Poisson process, or the amount of rainfall accumulated in a given period of time.

3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

````python
import math

def normal_pdf(x, mean, std):
    expt = -(x-mean)**2/(2*std**2)
    pdf = (1/(math.sqrt(2*math.pi) * std))*math.exp(expt)
    return pdf

x = 2.5
mean = 0
std = 1
pdf = normal_pdf(x, mean, std)
print(pdf)
````

4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

   The binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has the same probability of success. 
   
   The properties of the binomial distribution are -

   - The trials are independent.
   - Each trial has only two possible outcomes, success or failure.
   - The probability of success is the same for each trial.
   - The number of trials is fixed.
   - The random variable of interest is the number of successes.

   Two examples of events where binomial distribution can be applied are -

   - The number of people who respond "yes" to a survey question with a binary answer, such as "Do you own a car?"
   - The number of times a coin lands heads in a fixed number of tosses.

5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

n = 1000  # sample size
p = 0.4  # probability of success
s = np.random.binomial(n, p, size=n)  # generate the sample

plt.hist(s, bins=20, density=True, alpha=0.6)
plt.xlabel('Number of successes')
plt.ylabel('Probability density')
plt.title('Histogram of binomial distribution')
sns.kdeplot(s)
plt.show()
````

6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

````python
import math

def poisson_cdf(mean, k):
    cdf = 0
    for i in range(k + 1):
        cdf += math.exp(-mean) * mean**i / math.factorial(i)
    return cdf
````

The function `poisson_cdf(mean, k)` takes two arguments : `mean`, which is the mean of the Poisson distribution, and `k`, which is the point at which to calculate the cumulative distribution function.

10. In normal distribution with respect to mean position, where does the least frequent data appear?

- In a normal distribution, the least frequent data appear at the tails of the distribution, farthest away from the mean. 
- The normal distribution is symmetric around its mean, so the least frequent data are equally likely to appear on either side of the mean. 
- As you move further away from the mean, the probability density of the normal distribution decreases rapidly.
